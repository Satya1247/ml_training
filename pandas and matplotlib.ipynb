```python
#Matplotlib
import matplotlib.pyplot as plt
```

```python
days=[1,2,3,4,5,6,7]
insta_likes=[100,130,145,175,400,50,25]
plt.plot(days,insta_likes)
```

```python
plt.title("social media analysis")
plt.xlabel("number of days")
plt.ylabel("number of likes")
plt.plot(days,insta_likes,color="red",marker="o",markerfacecolor="black",linestyle="-.")
```

```python
facebook_likes=[111,234,257,334,445,500,600]
youtube_likes=[700,900,1100,1300,1500,2000,2500]
plt.title("social media comparision")
plt.xlabel("number of days")
plt.ylabel("number of likes")
plt.plot(days,insta_likes,color="blue",label="insta",marker="o")
plt.plot(days,facebook_likes,color="red",label="facebook",marker="o")
plt.plot(days,youtube_likes,color="orange",label="youtube",marker="o")
plt.legend(loc="upper left")
plt.grid(color="black",linewidth=2)
plt.xlim(1,5)
plt.ylim(1000,5000)
```
