```python
#concat and merge
df1=pd.DataFrame({'A':[1,2],'B':[3,4]})
df2=pd.DataFrame({'A':[5,6],'B':[7,8]})
df1
```

```python
df2
```

```python
df3=pd.concat([df1,df2])
df3
```

```python
#merge in panadas similar to the joins in sql
df1=pd.DataFrame({'A':[1,2],'B':[3,4]})
df2=pd.DataFrame({'A':[1,6],'B':[7,8]})
df3=pd.merge(df1,df2,on='A',how='right')
df3
```
